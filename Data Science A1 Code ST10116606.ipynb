```python
plt.figure(figsize=(8, 4))
plt.hist(df['track_popularity'], bins=10, range=(0, 100), color = 'yellow', edgecolor= 'black', alpha=0.7)
plt.title('Histogram of track popularity scores')
plt.xlabel('Popularity score')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['key'], bins=10, range=(0, 11), color = 'yellow', edgecolor= 'black', alpha=0.7)
plt.title('Histogram of key scores')
plt.xlabel('Key score')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show
```
